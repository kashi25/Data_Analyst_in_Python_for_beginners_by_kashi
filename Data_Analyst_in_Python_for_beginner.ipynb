```python
import pandas as pd
```

```python
# create a dataFrame
data = {'Name': ['kashi', 'kadlayan', 'Kabi', 'Kanindra'],
        'Age': [25, 34, 19, 34],
        'Place': ['Jhapa', 'Morang', 'Sunsari', 'Rukum']
}

df = pd.DataFrame(data)
```

```python
print(df)
```

```python
# Accessing the column in the row
print(df['Name'])
```

```python
# Accessing the first row
print(df.iloc[0])
```

```python
# accessing the second row
print(df.iloc[1])
```

```python
# accessing the first and second row
print(df.iloc[0:2])
```

```python
# data manipulation
df['Salary'] = [94900, 89077, 90766, 78900] #Adding new column
```

```python
print(df)
```

```python
df['Age'] = df['Age'] + 1 #increase the age by 1
print(df)
```

```python
df = df.drop('Place', axis=1)  #removing the 'Place' column
```

```python
print(df)
```

```python
print(df.describe()) #generate descriptive statistics
```

```python
# Data visualization
df.plot(kind = 'bar', x = 'Name', y = 'Salary')
```

```python
# data import/export
df.to_csv('datafram_data.csv', index = False) #export the dataFrame to csv file
```

Matplotlib is a popular data visualization library in Python that provides a wide range of functions and capabilities for creating static, animated, and interactive visualizations. Here's an introduction to plotting with Matplotlib:

```python
# pip install matplotlib
import matplotlib.pyplot as plt
```

```python
# data
x = [ 1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# create a line plot
plt.plot(x, y)

# add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')

# display the plot
plt.show()
```

Customizing Plots: Matplotlib provides numerous options for customizing the appearance of your plots. Some commonly used customization options include:

Changing line colors, styles, and markers.

Adding gridlines and legends.

Adjusting axes limits, ticks, and labels.

Setting plot backgrounds, titles, and text annotations.



```python
import matplotlib.pyplot as plt

# data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# creating the plots for both datasets

plt.plot(x, y1, label = 'Dataset 1')
plt.plot(x, y2, label = 'Dataset 2')


# adding legend, labels and title
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiplie line plots')

# display the plot
plt.show()
```

Saving Plots: You can save your Matplotlib plots as image files using the savefig function. Supported file formats include PNG, JPEG, PDF, and more

```python
import matplotlib.pyplot as plt

# data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]


# creating the plots for both datasets

plt.plot(x, y1, label = 'Dataset 1')
plt.plot(x, y2, label = 'Dataset 2')


# adding legend, labels and title
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiplie line plots')

# display the plot
plt.show()

# save the plot as an image file
plt.savefig('plot.png')
```

```python
# Adding Axis Labels and Title:
```

```python
import matplotlib.pyplot as plt

# Define data for the plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]


# create a plot
plt.plot(x, y)

# adding axis labels and title
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('plot Title')

# display the plot
plt.show()
```

```python
# mathplotlib
import matplotlib.pyplot as plt


# create a plot
plt.plot(x, y)

# set axis limit
plt.xlim(0,20)
plt.ylim(0,30)

# set custom tick value
plt.xticks([0, 5, 10])
plt.yticks([0,10, 20])

# display the plot
plt.show()
```

```python
# Adding Annotations and Text:
```

```python
import matplotlib.pyplot as plt


# create a plot
plt.plot(x, y)

# add text annotation
plt.text(2, 10, 'Annotation', fontsize = 12, color = 'red')

# display the plot
plt.show()
```
