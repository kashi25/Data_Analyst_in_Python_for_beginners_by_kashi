# Visualizing Qualitative Data
## Qualitative data is data that approximates and characterizes.
### it can be observed and recorded.
### it is  non numerical in nature also known as categorical data
### it's subtypes are nominal data and ordinal data
### Nominal data is data that has no inherent order.
### Ordinal data is data that falls into ordered categories.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings = pd.read_csv('E:/data analyst/listings.csv')
```

```python
listings
```

```python
sn.countplot(x = 'neighbourhood_group', data = listings)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings)
plt.show()
```

```python
# the black line in bargraph are called confidence interval
```

```python
# to remove the black line
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings, ci = False)
plt.show()
```

# Visualizing Quantitative Data

## Quantitative data is about numerical data.
## Data that consists of amount or quantities
## Two types of charts that are commonly used to visualize quantitative data
## histograms and Scatter plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
# here numpy is generally used as to visualized the quantitative data
```

```python
listings = pd.read_csv('E:/data analyst/listings.csv')
listings
```

```python
plt.hist(listings ['price'])
plt.xlabel('price(in Neplai RS)')
plt.show()
```

# note that the price are grouped into contiguous intervals called bins

```python
plt.hist(listings ['price'], bins = np.arange(0, 1100, 40))
plt.xlabel('price(in Neplai RS)')
plt.show()
```

```python
plt.scatter(x = listings ['price'], y = listings ['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = listings ['price'], y = listings ['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0, 1100)
plt.show()
```

```python
plt.scatter(x = listings ['price'], y = listings ['number_of_reviews'], s = 5)
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0, 1100)
plt.show()
```
